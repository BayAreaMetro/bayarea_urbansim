```python
import os
import sys
sys.path.insert(0, 'X:\\bayarea_urbansim\\scripts')
from baus_analytics_tools import *
```

```python
# parcel to TAZ/SD lookup
p_taz = pd.read_csv(r'M:\Data\Urban\BAUS\PBA50\p10_parcel_TAZ_SD.csv')

print(p_taz.shape)
display(p_taz.head())
```

```python
# location of two runs: run270-P, run269-NP
folder = 'X:\\bayarea_urbansim\\runs'

runs = {'run290': 'NP',
        'run292': 'P'}
```

```python
p_node = p_taz[['parcel_id']].copy()
for filename in os.listdir(folder):
    if ('run290_parcel_data_allCols' in filename) or ('run292_parcel_data_allCols' in filename):
        print(filename)
        scen = runs[filename.split('_')[0]]
        yr = filename.split('.')[0][-4:]
        print('scenario: ', scen, '; year: ', yr)
        
        df = pd.read_csv(os.path.join(folder, filename),
                         usecols = ['parcel_id', 'node_id', 'tmnode_id'])
        print('{} rows of parcel data'.format(df.shape[0]))
        df.rename(columns = {'node_id': 'node_id_'+yr+'_'+scen,
                             'tmnode_id': 'tmnode_id_'+yr+'_'+scen}, inplace=True)
        display(df.head())
        p_node = p_node.merge(df, on='parcel_id', how='outer')
```
