```python
import pandas as pd
import os
```

```python
# input/output directories
run_dir = 'X:\\bayarea_urbansim\\runs'
output_dir = r'M:\Data\Urban\BAUS\PBA50\explore_developer_model\runs_new_parcel_subzone_input'
```

```python
runs = {'run310': 'NP',
        'run313': 'P'}
```

```python
# read TAZ-SD number crosswalk
taz_f = pd.read_csv('M:/Data/GIS layers/TM1_taz/bayarea_rtaz1454_rev1_WGS84.csv',
                    usecols = ['TAZ1454', 'SUPERD'])
taz_f.drop_duplicates(inplace=True)

# read SD number-name crosswalk 
sd_name = pd.read_csv(r'X:\bayarea_urbansim\data\superdistricts.csv',
                      usecols=['number', 'name'])
sd_name.drop_duplicates(inplace=True)

# merge the two
TAZ_SD = taz_f.merge(sd_name, left_on='SUPERD', right_on='number', how='left')
TAZ_SD = TAZ_SD[['TAZ1454', 'SUPERD', 'name']]
TAZ_SD.rename(columns = {'TAZ1454': 'TAZ', 'name': 'SD_NAME'}, inplace=True)

display(TAZ_SD)
```

```python
# TAZ-level logsums

logsum_taz_all = pd.DataFrame(columns = ['TAZ', 'year',
                                         'zone_cml_p', 'zone_combo_logsum_p', 'zone_cnml_p',
                                         'zone_cml_np', 'zone_combo_logsum_np', 'zone_cnml_np'])

for filename in os.listdir(run_dir):
    if ('run310_taz_logsum' in filename) or ('run313_taz_logsum' in filename):
        print(filename)
        scen = runs[filename.split('_')[0]]
        year = filename.split('.')[0][-4:]
        print(scen, year)
    
        df = pd.read_csv(os.path.join(run_dir, filename))
        df['year'] = year
        
        if 'logsums_2010' in filename:
            df.rename(columns = {'zone_id': 'TAZ'}, inplace=True)
        
        if scen == 'P':
            df.rename(columns = {'zone_cml' : 'zone_cml_p',
                                 'zone_combo_logsum': 'zone_combo_logsum_p',
                                 'zone_cnml': 'zone_cnml_p'}, inplace=True)
        elif scen == 'NP':
            df.rename(columns = {'zone_cml' : 'zone_cml_np',
                                 'zone_combo_logsum': 'zone_combo_logsum_np',
                                 'zone_cnml': 'zone_cnml_np'}, inplace=True)
        display(df.head(2))
        print(df.shape[0])
        
        logsum_taz_all = pd.concat([logsum_taz_all, df])
        print(logsum_taz_all.shape[0])
```

```python
# drop NA rows
logsum_taz_all_p = logsum_taz_all[['TAZ', 'year', 'zone_cml_p', 'zone_combo_logsum_p', 'zone_cnml_p']]
logsum_taz_all_np = logsum_taz_all[['TAZ', 'year', 'zone_cml_np', 'zone_combo_logsum_np', 'zone_cnml_np']]

logsum_taz_all_p.dropna(subset=['zone_cml_p', 'zone_combo_logsum_p', 'zone_cnml_p'], inplace=True)
logsum_taz_all_np.dropna(subset=['zone_cml_np', 'zone_combo_logsum_np', 'zone_cnml_np'], inplace=True)

logsum_taz = logsum_taz_all_p.merge(logsum_taz_all_np, on=['TAZ', 'year'], how='outer')
```

```python
logsum_taz
```

```python
# add SD tag and export
logsum_taz = logsum_taz.merge(TAZ_SD, on='TAZ', how='left')

logsum_taz_2010_2025 = logsum_taz.loc[logsum_taz.year == '2010']
print(logsum_taz_2010_2025.shape[0])
logsum_taz_2010_2025.to_csv(os.path.join(output_dir, 'baus_logsum_taz_2010_2025.csv'), index=False)

logsum_taz_2030_2050 = logsum_taz.loc[logsum_taz.year == '2050']
print(logsum_taz_2030_2050.shape[0])
logsum_taz_2030_2050.to_csv(os.path.join(output_dir, 'baus_logsum_taz_2030_2050.csv'), index=False)
```

```python
# parcel level logsums

logsum_parcel_all = pd.DataFrame(columns = ['parcel_id', 'year',
                                            'cml_p', 'combo_logsum_p', 'cnml_p',
                                            'cml_np', 'combo_logsum_np', 'cnml_np'])

for filename in os.listdir(run_dir):
    if ('run310_parcel_logsum' in filename) or ('run313_parcel_logsum' in filename):
        print(filename)
        scen = runs[filename.split('_')[0]]
        year = filename.split('.')[0][-4:]
        print(scen, year)
    
        df = pd.read_csv(os.path.join(run_dir, filename))
        df['year'] = year

        if scen == 'P':
            df.rename(columns = {'cml' : 'cml_p',
                                 'combo_logsum': 'combo_logsum_p',
                                 'cnml': 'cnml_p'}, inplace=True)
        elif scen == 'NP':
            df.rename(columns = {'cml' : 'cml_np',
                                 'combo_logsum': 'combo_logsum_np',
                                 'cnml': 'cnml_np'}, inplace=True)
#         display(df.head(2))
        print(df.shape[0])
        
        logsum_parcel_all = pd.concat([logsum_parcel_all, df])
        print(logsum_parcel_all.shape[0])
```

```python
logsum_parcel_all
```

```python
# drop NA rows
logsum_parcel_p = logsum_parcel_all[['parcel_id', 'year', 'cml_p', 'combo_logsum_p', 'cnml_p']]
logsum_parcel_np = logsum_parcel_all[['parcel_id', 'year', 'cml_np', 'combo_logsum_np', 'cnml_np']]

logsum_parcel_p.dropna(subset=['cml_p', 'combo_logsum_p', 'cnml_p'], inplace=True)
logsum_parcel_np.dropna(subset=['cml_np', 'combo_logsum_np', 'cnml_np'], inplace=True)

logsum_parcel = logsum_parcel_p.merge(logsum_parcel_np, on=['parcel_id', 'year'], how='outer')
```

```python
logsum_parcel
```

```python
# read parcel_zone crosswalk
pt = pd.read_csv(r'X:\bayarea_urbansim\data\2018_10_17_parcel_to_taz1454sub.csv', 
                 usecols=['PARCEL_ID', 'ZONE_ID'])
pt['PARCEL_ID'] = pt['PARCEL_ID'].apply(lambda x: int(round(x)))
pt.rename(columns={'ZONE_ID': 'TAZ',
                   'PARCEL_ID': 'parcel_id'}, inplace=True)

p_taz = pt.merge(TAZ_SD, on='TAZ', how='left')
print(p_taz.shape)
display(p_taz.head())
```
