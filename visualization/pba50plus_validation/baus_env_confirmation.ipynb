This is just a quick comparison to confirm that the new BAUS Python environments on `lumodel2` and `lumodel3` are working correctly.

The reference outputs come from `satmodel3` which uses the previous (standard) BAUS Python environment. I (Drew) ran v0 and v1 on `satmodel3` so these are good comparisons against the corresponding runs on `lumodel2` (v0) and `lumodel3` (v1).

```python
import pandas as pd
from pathlib import Path

satmodel3_v0 = Path(r'\\satmodel3\SatModel3Share\baus_main_current_PBA50_inputs\outputs\pba50_fbp_pr319_v0\travel_model_summaries')
satmodel3_v1 = Path(r'\\satmodel3\SatModel3Share\baus_main_current_PBA50_inputs\outputs\pba50_fbp_pr319_v1\travel_model_summaries')
lumodel2_v0 = Path(r'\\lumodel2\LUModel2Share\baus_main_current_PBA50_inputs\outputs\pba50_fbp_pr319_v0_lumodel2\travel_model_summaries')
lumodel3_v1 = Path(r'\\lumodel3\LUModel3Share\baus_main_current_PBA50_inputs\outputs\pba50_fbp_pr319_v1\travel_model_summaries')
```

```python
# v0 - satmodel3 vs lumodel2
# Not identical but quite close, close enough for me! - DSL

for year in [2015, 2035, 2050]:
    print('##########', year, '##########')
    for source in [satmodel3_v0, lumodel2_v0]:
        print(source)
        df = pd.read_csv(source / f'taz1_summary_{year}.csv')
        print(df.groupby('COUNTY')[['TOTHH', 'TOTEMP']].sum() / df[['TOTHH', 'TOTEMP']].sum())
        print()
```

```python
# v1 - satmodel3 vs lumodel3
# Again not identical but plenty close

for year in [2015, 2035, 2050]:
    print('##########', year, '##########')
    for source in [satmodel3_v1, lumodel3_v1]:
        print(source)
        df = pd.read_csv(source / f'taz1_summary_{year}.csv')
        print(df.groupby('COUNTY')[['TOTHH', 'TOTEMP']].sum() / df[['TOTHH', 'TOTEMP']].sum())
        print()
```
